## Quick checks

Make sure the GeoJSON exists (created by `pyagri-extract`):

```bash
ls -l data/harvest_fields.geojson
```

````python
import geopandas as gpd
import matplotlib.pyplot as plt

gdf = gpd.read_file('data/harvest_fields.geojson')
print('Features:', len(gdf))
gdf.head()
````

````python
# Simple plot colored by Year (if available)
fig, ax = plt.subplots(figsize=(10, 8))
if 'Year' in gdf.columns:
    gdf.plot(column='Year', ax=ax, cmap='tab20', legend=True, edgecolor='black')
else:
    gdf.plot(ax=ax, color='lightgreen', edgecolor='black')
ax.set_title('Harvest Fields')
plt.show()
````
